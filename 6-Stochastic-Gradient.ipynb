```python
import numpy as np
```

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Target Output
y = np.array([0, 1, 1, 1])

print("Input Dataset:")
for i in range(len(X)):
    print(f"Input: {X[i]} --> Output: {y[i]}")
```

```python
np.random.seed(42)
weights = np.random.randn(2) * 0.01
bias = 0.0

learning_rate = 0.1
epochs = 50

print("\nInitial Weights:", weights)
print("Initial Bias:", bias)
```

```python
def activation_function(x):
    return 1 if x >= 0 else 0
```

```python
for epoch in range(epochs):
    total_error = 0

    for i in range(len(X)):
        # Calculate weighted sum
        net_input = np.dot(X[i], weights) + bias

        # Prediction
        prediction = activation_function(net_input)

        # Error
        error = y[i] - prediction
        total_error += abs(error)

        # SGD Weight Update Rule
        weights = weights + learning_rate * error * X[i]
        bias = bias + learning_rate * error

    # Print progress every 5 epochs
    if (epoch + 1) % 5 == 0:
        print(f"Epoch {epoch+1}, Total Error: {total_error}")
```

```python
print("\nTraining Completed")
print("Final Weights:", weights)
print("Final Bias:", bias)
```

```python
print("\nFinal Predictions after Training:")

correct = 0

for i in range(len(X)):
    net_input = np.dot(X[i], weights) + bias
    prediction = activation_function(net_input)

    print(f"Input: {X[i]} --> Actual: {y[i]}, Predicted: {prediction}")

    if prediction == y[i]:
        correct += 1
```

```python
accuracy = (correct / len(X)) * 100
print(f"\nImproved Prediction Accuracy: {accuracy:.2f}%")
```
