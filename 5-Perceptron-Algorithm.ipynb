```python
import numpy as np
```

```python
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

y = np.array([0, 1, 1, 1])
```

```python
weights = np.zeros(2)
bias = 0
learning_rate = 0.1
epochs = 10
```

```python
def activation_function(x):
    if x >= 0:
        return 1
    return 0
```

```python
for epoch in range(epochs):
    print(f"Epoch {epoch + 1}")

    for i in range(len(X)):
        # Calculate weighted sum
        net_input = np.dot(X[i], weights) + bias

        # Get prediction
        prediction = activation_function(net_input)

        # Calculate error
        error = y[i] - prediction

        # Update weights and bias
        weights = weights + learning_rate * error * X[i]
        bias = bias + learning_rate * error

        print(f"Input: {X[i]}, Target: {y[i]}, Prediction: {prediction}, Error: {error}")

    print(f"Updated Weights: {weights}, Bias: {bias}")
    print("-" * 50)
```

```python
print("\nFinal Output after Training:")

for i in range(len(X)):
    net_input = np.dot(X[i], weights) + bias
    prediction = activation_function(net_input)

    print(f"Input: {X[i]} --> Predicted Output: {prediction}")
```
