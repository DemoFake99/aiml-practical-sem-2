```python
# 1. Create array of n element

arr = np.array([10,20,30,40,50,60,70,80, 99, 100])
```

```python

# 2. Write a statement to print the first element

print(arr[0])
```

```python
# 3. Write a statement to print the element at index 3

print(arr[3])
```

```python
# 4. Write a statement to print the ast element using positive indexing

print(arr[arr.size - 1])
```

```python
# 5. Write a statement to print the last element using negative indexing

print(arr[-1])
```

```python
# 6. What will be the output of arr[4]

print(arr[4])
```

```python
# 7. Write a statement to print first 3 element

print(arr[:3])
```

```python
# 8. Write a statement to print element from index 2 to 5(not including 5)

print(arr[2:5])
```

```python
# 9. Write a statement to print element from index 3 to end

print(arr[3:])
```

```python
# 10. Write a statement to print all element except the 1st one

print(arr[1:])
```

```python
# 11. What is the output of arr[:4]

print(arr[:4])
```

```python
# 12. Write a statement to print every 2nd element

print(arr[::2])
```

```python
# 13. Write a statement to print every 3rd element

print(arr[::3])
```

```python
# 14. Write a statement to print element from index 1 to 5 with step 2

print(arr[1:5:2])
```

```python
# 15. Write a statement to reverse the array

print(arr[::-1])
```

```python
# 16. Write a statement to print every 2nd element in reverse order

print(arr[::-2])
```

```python
# 17. Write a statement to print elements which are greater than 30

print(arr[arr >= 30])
```

```python
# 18. Write a statement to print even elements

print(arr[arr % 2 == 0])
print(arr[arr % 2 != 0])
```

```python
# 19. Write a statement to print element divisible by 3

print(arr[arr % 3 == 0])
```
