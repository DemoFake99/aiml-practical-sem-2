```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('Data/iris.csv')
data.head()
```

```python
sns.scatterplot(data=data, x='PetalLengthCm', y='PetalWidthCm', hue='Species', palette='viridis')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
y_prediction = kmeans.fit_predict(data[['PetalLengthCm', 'PetalWidthCm']])
```

```python
sns.scatterplot(
    data=data,
    x='PetalLengthCm',
    y='PetalWidthCm',
    hue='Species',
    palette='viridis'
)

centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=200, marker='X')

plt.show()
```
